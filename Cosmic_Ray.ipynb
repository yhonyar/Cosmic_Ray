## 1. Introducción 


El observatorio Pierre Auger se encuentra ubicado en la ciudad de Malargue (Mendoza - Argentina). Esta es una iniciativa en la que participan 100 instituciones de 18 países para estudiar las partículas de alta energía que provienen del espacio exterior, es decir los rayos cósmicos (RC). 

image.png

Estos RC generan cascadas de partículas en la atmósfera que pueden ser detectadas en la superficie terrestre, así se dispone de diversos tipos de detectores como se muestra a continuación:  

•	Detectores Cherenkov: Los cuales son unos tanques de agua para detectar RC por la emisión de radiación Cherenkov.  
•	Detectores de Fluorescencia: Esta es la detección de fluorescencia debido a la desexitación en las moleculas atmosféricas después de haber sido excitadas por la cascada.  
•	Antenas de Radio: La cascada también produce radiaciones de baja frecuencia en el rango de las señales de radio, por ello se pueden colocar antenas en tierra que rastreen la distribución espacial, polarización e intensidad de estas.  
•	Planos Centelleadores: Estos elementos se ubican bajo tierra a fin de que la radiación sea absorbida y solo queden los muones a detectar.

El problema que surge en el análisis de las lecturas de estos detectores es que la cantidad de datos es significativamente grande, por lo cual surge la necesidad de automatizar los procesos mediante el uso del aprendizaje de máquina o Maching Learning (ML) entra a tallar.  
En particular, se requiere ML para lograr que el software clasifique y/o identifique el tipo de partícula primaria que produce la cascada detectada.

## 2. Marco Teórico

**Rayos Cósmicos**

Los rayos cósmicos (RC) representan un problema abierto en astrofísica, dado que sus mecanismos de generación involucran fenómenos de física de altas energías. Estos son partículas conformadas principalmente por protones (86%), partículas alfa (11%), electrones (2%) y núcleos atómicos pesados (1%), los cuales bombardean la tierra y al interactuar con la atmósfera  generan las denominadas cascadas atmosféricas extendidas (EAS).    
Las EAS tienen dos componentes principales, los cuales son la electromagnética y hadrónica. 

image.png

Así, esta cascada posee parámetros que se pueden medir a fin de correlacionar estos con el tipo de partícula primaria. Para el trabajo en cuestión, los parámetros fueron los siguientes:  
$X_{max}$ : Profundidad máxima a la que se desarrolla la cascada; su valor será menor mientras más pesado sea la partícula primaria.  
$Energy$: Energía de la partícula primaria, la cual es del orden de $10^{17} eV$.  
$Zenith$  y $Azimuth$: Son las orientaciones del eje de la cascada.  
$α$ o $alpha$: Es el ángulo magnético, que se define como el ángulo entre el eje de la cascada y la dirección del campo magnético terrestre en el lugar de observación.  
$antenas$: Representa el número de antenas colocadas en cada tanque Cherenkov para recolectar la señal de radio de la cascada.  
$muons$: Es la cantidad de muones detectados además de la cascada en estudio.  
$S_{rad}$: Es la energía radiada.  
$target$: Es el identificador de la partícula primaria, se define como $0$ para el Protón y $1$ para el Hierro.

**Machine Learning en Astrofísica de Altas Energías**

El aprendizaje automático (machine learning, en inglés) ha demostrado ser una herramienta valiosa en la astrofísica de altas energías. Una de las aplicaciones es utilizar para clasificar y detectar fuentes astrofísicas de alta energía en grandes conjuntos de datos, como datos de telescopios espaciales o detectores de partículas. Los algoritmos de aprendizaje automático pueden identificar patrones complejos en los datos que podrían indicar la presencia de fuentes astrofísicas, ayudando así a los astrónomos a descubrir y caracterizar nuevas fuentes.

a. Suporrt Vector Machine (SVM): El objetivo del algoritmo SVM es encontrar un hiperplano que separe de la mejor forma posible dos clases diferentes de puntos de datos. “De la mejor forma posible” implica el hiperplano con el margen más amplio entre las dos clases, representado por los signos más y menos en la siguiente figura. El margen se define como la anchura máxima de la región paralela al hiperplano que no tiene puntos de datos interiores. El algoritmo solo puede encontrar este hiperplano en problemas que permiten separación lineal; en la mayoría de los problemas prácticos, el algoritmo maximiza el margen flexible permitiendo un pequeño número de clasificaciones erróneas.

![SVM](./SVM.jpg)

b. Decision Tree (DT): Se trata de un diagrama de flujo que empieza con una idea principal y luego se ramifica según las consecuencias de tus decisiones. Se denomina “árbol de decisiones” por su semejanza con un árbol con muchas ramas.

![DT](./DT.png)

c. Logistic Regression: El método de regresión logística es un método estadístico que se usa para resolver problemas de clasificación binaria, donde el resultado solo puede ser de naturaleza dicotómica, o sea, solo puede tomar dos valores posibles. Por ejemplo, se puede utilizar para detectar la probabilidad que ocurra un evento.

![DT](./LR.png)

```python
import pandas as pd
import numpy as np
import math
import random

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV

import seaborn as sns
sns.set(style="whitegrid")
sns.set_theme()
import matplotlib.pyplot as plt
%matplotlib inline

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

Realizamos la lectura de los datos obtenido del observatorio Pierre Auger, en el cual tenemos 8 variables y un dato de salida

```python
cosmic_data = pd.read_csv("data_PFe_prepared_MC.txt", sep = " ")
cosmic_data.tail()
```

Para poder identificar las particulas depositdas en el target 0 o 1, separamos los datos como se muestra en el siguiente script

```python
cosmic_data0=cosmic_data[cosmic_data.target==0]
cosmic_data1=cosmic_data[cosmic_data.target==1]
```

Ahora procedemos a graficar de par en par las variables para poder observar cuales son las variables que nos ayudan a clasificar si es proton o hierro, es decir, si se ha depositado en el target 0 o 1.]

```python
fig = plt.figure(figsize=(17,30))
fig.clf()

a = 7
b = 4
n = -1

axes = fig.subplots(a,b)

for i in range(0,8,1):
    for j in range(i,8,1):
        if i != j:
            n = n + 1
            x_i = cosmic_data0.columns[i]
            x_j = cosmic_data0.columns[j]
            
            if n%b != 0:
                n1 = int(n/b)
                n2 = n - int(n/b)*b-1
                sns.scatterplot(ax = axes[n1,n2], data = cosmic_data0, y = cosmic_data0[x_j], x = cosmic_data0[x_i], 
                                label = "Target 0", color= 'blue')
                sns.scatterplot(ax = axes[n1,n2], data = cosmic_data1, y = cosmic_data1[x_j], x = cosmic_data1[x_i], 
                                label = "Target 1", color= 'red')
                axes[n1,n2].set_xlabel(x_i)
                axes[n1,n2].set_ylabel(x_j)
                axes[n1,n2].legend()
            else:
                n3 = int(n/b)-1
                n4 = b-1
                sns.scatterplot(ax = axes[n3,n4], data = cosmic_data0, y = cosmic_data0[x_j], x = cosmic_data0[x_i], 
                                label = "Target 0", color= 'blue')
                sns.scatterplot(ax = axes[n3,n4], data = cosmic_data1, y = cosmic_data1[x_j], x = cosmic_data1[x_i], 
                                label = "Target 1", color= 'red')
                axes[n3,n4].set_xlabel(x_i)
                axes[n3,n4].set_ylabel(x_j)
                axes[n3,n4].legend() 
```

Segun las graficas que se muestran, las variables que nos ayudan a diferenciar entre protones y hierro son la profundidad máxima (Xmax), la energía depositada (energy), el número de muones detectados (muons) y el ángulo magnético (alpha). Entonces, ahora vamos a modificar nuestro DataFrame inicial y solo considerar las variables mencionadas.

```python
cosmic_data_mod = cosmic_data[['Xmax','energy','muons','alpha','target']]
cosmic_data_mod.head()
```

Una vez seleccionada las variables que se usarán, ahora procederemos a usar los métodos de ML supervisado como DT, SVM y LR, con el fin de compara cual de los metodos mencionados nos permiten obtener resultados con una mayor precisión, asimismo, aremos la busqueda de la población de datos optimos que nos permita obtener una precisión mayor para de esta forma, en las siguientes pruebas, use ese dato con el fin de reducir el costo computacional.

```python
y_data = cosmic_data_mod['target']
x_data = cosmic_data_mod.drop('target',axis = 1)
num = 10
num_div = 99

sc = np.zeros((num_div*num,4))

for l in range(0,num,1):
    #Matriz de presición de los modelos SVM, DT y LR
    for n in range(0,num_div,1):
        
        m = random.uniform(0.01,0.99)
        x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x_data, y_data, test_size = m)
        sc[n+l*num_div,0] = m
        
        #Usando el modelo SVM
        model = svm.SVC()
        model.fit(x_training_data, y_training_data)
        predictions1 = model.predict(x_test_data)
        sc[n+l*num_div,1] = metrics.accuracy_score(predictions1,y_test_data)

        #Usando el modelo Arbol de desiciones
        modelo2 = tree.DecisionTreeClassifier()
        modelo2.fit(x_training_data, y_training_data)
        predictions2 = modelo2.predict(x_test_data)
        sc[n+l*num_div,2] = metrics.accuracy_score(predictions2,y_test_data)

        #Logictic Regression 
        modelLR = LogisticRegression()
        modelLR.fit(x_training_data, y_training_data)
        predictions3 = modelLR.predict(x_test_data)
        sc[n+l*num_div,3] = metrics.accuracy_score(predictions3,y_test_data)
```

En el siguiente entrada creamos un nuevo DataFrame que almacena la precision con el que el modelo ajusta los datos en relación el el tamaño de muestra usada para la prueba del modelo, como se muestra en la siguiente salida.

```python
Modelos = pd.DataFrame(sc, columns=['Tamaño de muestra','Support Vector Machine','Decision Tree','Logistic Regression'])
Modelos.describe()
```

La siguiente entrada es para comparar graficamente los modelos en función de su presición, como se muestra en la siguiente salida.]

```python
eje_x = np.linspace(0,1)

poly1, cov1 = np.polyfit(Modelos["Tamaño de muestra"], Modelos["Decision Tree"], deg=2, cov = True)

sns.scatterplot(x = 'Tamaño de muestra', y = 'Decision Tree', data = Modelos, label='DT')
sns.scatterplot(x = 'Tamaño de muestra', y = 'Logistic Regression', data = Modelos, label='LR')
sns.scatterplot(x = 'Tamaño de muestra', y = 'Support Vector Machine', data = Modelos, label='SVM')

plt.plot(eje_x, np.polyval(poly1, eje_x), '-k')
plt.xlabel('Tamaño de muestra')
plt.ylabel('Presición')
plt.title('Comparación de los modelos en función de su presición')
plt.legend()
plt.show()
```

Obtenemos el tamaño de muestra óptimo

```python
a = poly1[0]
b = poly1[1]

sa = math.sqrt(cov1[0,0])
sb = math.sqrt(cov1[1,1])
Cab=cov1[1,0]

optimo = -b/(2*a)

error_op = 1/(2*a**2)*math.sqrt((a*sb)**2 + (b*sa)**2-2*a*b*Cab)
```

```python
print("El tamaño de muestro óptimo es: ", optimo, "±", error_op)
```

Sabiendo el valor óptimo para la prueba del modelo, asi mismo, teniendo en claro que el mejor metodo es el DT, en ese sentido, realizamos la prueba y el entrenamiento con este modelo y con el valor optimo calculado.

```python
x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x_data, y_data, test_size = optimo)
```

```python
modelo2 = tree.DecisionTreeClassifier()
modelo2.fit(x_training_data, y_training_data)
predictions2 = modelo2.predict(x_test_data)

sc3 = metrics.accuracy_score(predictions2,y_test_data)

report2 = classification_report(y_test_data, predictions2)
print(report2)
```

Podemos ver claramente en la matriz de confusión que el modelo DT del total de datos de prueba, 251, los valore de verdadero postivo y verdadero negativo muestras valores de 109 y 113 lo cual nos indica que nuestro modelo de DT tiene una precisión de un 89 y 88 %, respectivamente.

```python
mat = confusion_matrix(y_test_data, predictions2)
plt.subplots(figsize=(3,3))
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, linewidth=.5)
plt.xlabel('Valores reales')
plt.ylabel('Valores pronosticados por el modelo')
```
